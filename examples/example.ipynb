```python
import pandas as pd
from src.cleaning import exploratory_descriptive_analysis
import seaborn as sns

df = pd.read_csv("nutrition.csv")

eda = exploratory_descriptive_analysis(df)
eda
variables = list(df.columns)
variables.remove('label')
print(variables)
```

```python
histogram(df, "Hisotgram", variables, "", "", edgecolor='black', bins=12, color=None, density=True, number=8, figsize=(30,18), row=3, column=2)
```

```python
histogram_kde(df, "Histogram", ["calories", "protein", "sugars", "sodium"], "calories", "count", edgecolor='black', bins=30, color=None, kde=True, number=4, figsize=(30,18), row=2, column=2)
```

```python
df_numeric = df.select_dtypes(include=["number"])
df_corr = df_numeric.corr(method = "pearson")

plt.figure(figsize=(20,12))
sns.heatmap(df_corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
